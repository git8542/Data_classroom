```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
learning_rate = .1
decay_rate = .9
decay_steps = 50
global_steps = 3000
```

```python
def construct_graph(learning_rate,decay_steps, decay_rate, staircase):
    
    W = tf.Variable(tf.truncated_normal(shape=(2,2),dtype=tf.float32))
    l2_loss = tf.nn.l2_loss(W)
    
    global_step = tf.Variable(0, trainable=False)
    
    learning_rate = tf.train.exponential_decay(learning_rate=learning_rate,
                                              global_step=global_step,
                                              decay_steps=decay_step,
                                              decay_rate=decay_rate,
                                              staircase=staircase)
    optimizer = tf.train.GradientDescentOptimizer(learning_rate=learning_rate)
    train = optimizer.minimize(loss=l2_loss,global_step=global_step)
    initializer = tf.global_variables_initializer()
    
    return (initializer, train, global_step, learning_rate)
```

```python
(initializer, train, global_step, learning_rate) = construct_graph(learning_rate, decay_steps, decay_rate, staircase=False)
```

```python
X_continuous = []
Y_continuous = []
with tf.Session() as sess:
    sess.run(initializer)
    global_step_, learning_rate_= sess.run(fetches=[global_step, learning_rate])
    X_continuous.append(global_step_ / decay_steps)
    Y_continuous.append(learning_rate_)
    print("global step: %d, learning rate: %f" %(global_step_, learning_rate_))
```

```python
(initializer, train, global_step, learning_rate) = \
    construct_graph(learning_rate, decay_steps, decay_rate, staircase=True)
```

```python
X_staircase = []
Y_staircase = []
with tf.Session() as sess:
    sess.run(initializer)
    for _ in range(global_steps):
        sess.run(train)
        global_step_,learning_rate_ = sess.run(fetches=[global_step, learning_rate])
        X_staircase.append(global_step_ / decay_steps)
        Y_staircase.append(learning_rate_)
        print("global step: %d, learning_rate:%f" %(global_step_, learning_rate_))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
curve_continuous = ax.plot(X_continuous,Y_continuous,"r", label="continuous learning rate")
curve_staircase = ax.plot(X_staircase, Y_staircase,"b", label="staircase learning rate")
ax.legend()
```
