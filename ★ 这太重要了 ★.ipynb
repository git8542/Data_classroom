```python
coke1 = pd.read_csv("35.csv")
coke1 = coke1.drop(["合约", "ROWNUM"], axis=1)
from sklearn.cluster import KMeans
clf = KMeans(n_clusters=3)

pre = clf.fit_predict(coke1)
pre
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
newdata = pca.fit_transform(coke1)
print(newdata[:4])
x1 = [x[0] for x in newdata]
x2 = [x[1] for x in newdata]
```

```python
plt.scatter(x1, x2, c=pre, marker="x")
plt.xlabel("x features")
plt.ylabel("y features")
plt.savefig("bankloan.png",dpi=1000)
plt.show()
```
