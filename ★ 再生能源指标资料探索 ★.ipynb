# 再生能源指标资料探索

```python
import pandas as pd
```

```python
def get_energy():
    
    """
    clean the energy data
    """
    energy = pd.read_excel('https://storage.googleapis.com/um_ds_intro/Energy%20Indicators.xls', skiprows=18, skipfooter=38, header=None, na_values='...')
    energy = energy.drop([0,1], axis=1)
    energy = energy.rename(columns={2: "Country", 3:"Energy Supply",
                                   4:"Energy Supply per Capita", 5:"% Renewable"})
    energy["Energy Supply"] = energy["Energy Supply"] * 1000000
    energy["Country"] = energy["Country"].str.replace(r"\s\(.*\)", "")

    energy["Country"] = energy["Country"].str.replace(r"[0-9]+", "")

    change_index = energy[energy["Country"].isin(["Republic of Korea",
                                                 "United States of America",
                                                 "United Kingdom of Great Britain and Northern Ireland",
                                                 "China, Hong Kong Special Administrative Region"])].index
    name_to_change = ["Hong Kong", "South Korea", "United Kingdom",
                     "United States"]

    for i, c_name in zip(change_index, name_to_change):
        energy.at[i, "Country"] = c_name
    energy = energy.set_index("Country")
    return energy
```

```python
def get_GDP():
    """
    clean GDP data
    
    """
    import pandas as pd 
    GDP = pd.read_csv('https://storage.googleapis.com/um_ds_intro/world_bank.csv', skiprows=4)
    GDP = pd.concat([GDP.loc[:, "Country Name"], GDP.loc[:, "2006":"2015"]], axis=1)

    change_index = GDP[GDP["Country Name"].isin(["Korea, Rep.", "Iran, Islamic Rep.", "Hong Kong SAR, China"])].index
    name_to_change = ["Hong Kong", "Iran", "South Korea"]
    for i , c_name in zip(change_index, name_to_change):
         GDP.at[i, "Country Name"] = c_name  # alter country name
    GDP = GDP.set_index("Country Name")
    return GDP   
```

```python
def get_ScimEn():
    """
    clean ScimEn data 
    """
    import pandas as pd
    ScimEn = pd.read_excel('https://storage.googleapis.com/um_ds_intro/scimagojr-3.xlsx')
    ScimEn = ScimEn.iloc[:15, :].set_index("Country")
    return ScimEn
```

```python
def answer_one():
    energy = get_energy()
    GDP = get_GDP()
    ScimEn = get_ScimEn()
    df = ScimEn.join(energy)
    df = df.join(GDP)
    return df
```

```python
answer_one()
```

```python
def answer_two():
    energy = get_energy()
    GDP = get_GDP()
    ScimEn = get_ScimEn()
    df_outer = ScimEn.join(energy, how="outer")
    df_outer = df_outer.join(GDP, how="outer")
#     df_inner = answer_one()
    ans = df_outer.shape[0]
    return ans

answer_two()
```

```python
# caculate GDP of Top15 Country during the past decade
def answer_three():
    Top15 = answer_one()
    avgGDP = Top15.loc[:, "2006":"2015"].mean(axis=1).sort_values(ascending=False)
    return avgGDP

answer_three()
```

```python
def answer_four():
    Top15 = answer_one()
    ranked_6th_country = answer_three().index[5]
    ranked_6th_2015 = Top15.loc[ranked_6th_country, "2015"]
    ranked_6th_2006 = Top15.loc[ranked_6th_country, "2006"]
    return ranked_6th_2015 - ranked_6th_2006

answer_four()
```

```python
def answer_five():
    Top15 = answer_one()
    ans = Top15["Energy Supply per Capita"].mean()
    return ans

answer_five()
```

```python
def answer_six():
    Top15 = answer_one()
    max_idx = Top15["% Renewable"].idxmax()
    max_val = Top15["% Renewable"].max()
    return max_idx, max_val

answer_six()
```

```python
def answer_seven():
    Top15 = answer_one()
    Top15["Self-crtations Ratio"] = Top15["Self-citations"] / Top15["Citations"]
    max_idx = Top15["Self-citations Ratio"].idxmax()
    max_val = Top15["Self-citations Ratio"].max()
    return max_idx, max_val

answer_seven()
```

```python
def answer_eight():
    Top15 = answer_one()
    Top15["Population"] = Top15["Energy Supply"] / Top15["Energy Supply per Capita"]
    Top15 = Top15.sort_values("Population", ascending= False)
    ranked_3rd = Top15.index[2]
    return ranked_3rd

answer_eight()
```

```python
def answer_nine():
    Top15 = answer_one()
    Top15["Population"] = Top15["Energy Supply"] / Top15["Energy Supply per Cpaita"]
    Top15["Citable documents per capita"] = Top15["Citable documents"] / Top15["Population"]
    ans = Top15["Citable documents per capita"].corr(Top15["Energy Supply per Capita"])
    return ans

answer_nine()
```

```python
def answer_ten():
    Top15 = answer_one()
    median_val = Top15["% Renewable"].median()
    HighRenew = Top15["% Renewable"].map(lambda x: 1 if x >= median_val else 0)
    return HighRenew

answer_ten()
```

```python
ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}

def answer_eleven():
    Top15 = answer_one()
    Top15["Population"] = Top15["Energy Supply"] / Top15["Energy Supply per Capita"]
    Top15 = Top15.reset_index()
    Top15["Continent"] = Top15["Country"].map(ContinentDict)
    
    pop_size = Top15.groupby("Continent")["Population"].count()
    pop_sum = Top15.groupby("Continent")["Population"].sum()
    pop_mean = Top15.groupby("Continent")["Population"].mean()
    pop_std = Top15.groupby("Continent")["Population"].std()
    
    df = pd.concat([pop_size, pop_sum, pop_mean, pop_std], axis=1)
    df.columns = ["size", "sum", "mean", "std"]
    return df

answer_eleven()
```

```python
def answer_twelve():
    Top15 = answer_one()
    Top15 = Top15.reset_index()
    Top15["% Renewable Bin"] = pd.cut(Top15["% Renewable"], bins=5,
                                     labels =["lowest", "low", "medium", "high", "highest"])
    Top15["Continent"] = Top15["Country"].map(ContinentDict)
    ans = Top15.groupby(["Continent", "% Renewable Bin"])["Country"].mean()
    return ans

answer_twelve()
```

```python
def answer_thirteen():
    Top15 = answer_one()
    Top15["PopEst"] = Top15["Energy Supply"] / Top15["Energy Supply per Capita"]
    Top15["PopEst"] = Top15["PopEst"].map("{:,}".format)
    return Top15[["PopEst"]]

answer_thirteen()
```
